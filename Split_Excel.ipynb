The objective is to create excel files for each city in excel 
and to automate this process for future use

```python
import pandas as pd
```

```python
df_pth = R"E:\Downloads\Phonebook 1.xlsx"
```

```python
df = pd.read_excel(df_pth)
```

```python
df.head()
```

```python
df.info()
```

```python
#no of unique cities
city_lst = df['city'].unique()
len(city_lst)
```

```python
# looping on city lst and filter the data by city then saving this data in a separete excel file
for city in city_lst:
    print(f"City name:"+city)
    city_df= df[df['city']==city]
    file_name= city + ".xlsx"
    city_df.to_excel(file_name)
    
```
